```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
data=pd.read_csv("data_sets/data1/cleansed.csv")
```

```python
data.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
data
```

# Binary Encoding

```python
data['TypeofContact'].unique()
```

```python
def type_of_contact_binary(a):
    if a=='Self Enquiry':
        return 1
    elif a=='Company Invited':
        return 0
```

```python
data['TypeofContact']=data['TypeofContact'].apply(type_of_contact_binary)
```

```python
data
```

```python
data.info()
```

```python
sns.countplot(data['TypeofContact'])
```

# Label Encoding

```python
sns.countplot(data['Occupation'])
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
data['Occupation']=le.fit_transform(data['Occupation'])
```

```python
sns.countplot(data['Occupation'])
```

# One Hot Encoding

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe=OneHotEncoder()
df=ohe.fit_transform(data[['Gender']]).toarray()
```

```python
df=pd.DataFrame(df)
```

```python
data=pd.concat([data,df],axis=1)
```

```python
data
```

```python
data=pd.get_dummies(data,columns=["Designation"])
```

```python
data
```

# Ordinal Encoding

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
sns.countplot(data['MaritalStatus'])
```

```python
oe=OrdinalEncoder()
data['MaritalStatus_enc']=oe.fit_transform(data[['MaritalStatus']])
```

```python
data['MaritalStatus_enc']
```

```python
sns.countplot(data['MaritalStatus_enc'])
```

# Hash Encoding

```python
from sklearn.feature_extraction import FeatureHasher
```

```python
sns.countplot(data['ProductPitched'])
```

```python
he=FeatureHasher(n_features=2,input_type="string")
hashed=he.fit_transform(data['ProductPitched']).toarray()
```

```python
hashed_df=pd.DataFrame(hashed)
```

```python
data=pd.concat([data,hashed_df],axis=1)
```
